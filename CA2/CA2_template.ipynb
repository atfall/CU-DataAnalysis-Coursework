```python
import numpy as np
from scipy.stats import uniform, norm
from tqdm import tqdm
%matplotlib inline


# define a function for the probability
# prob(posterior) = normal(likelihood) x normal(prior)
def posterior(theta_mass, theta_distance): 
    data=owf['strain']
    #X_hat=np.mean(data)
    n = len(data)
    #var = np.var(data, ddof=1)
    sig = np.mean(abs(owf['strain'][cutoff:-1]))
    #mu_0 = np.random.uniform(20,100)
    #s_0 = 5.
    log_likelihood = (-0.5)*np.sum(((owf['strain'] - scaler(theta_mass, theta_distance, owf['shifted_time'])[1])**2)/(sig**2))
    val = log_likelihood#*norm.pdf(theta,mu_0,s_0)
    return val


# set up MCMC step paramaters
N_mcmc = 1_000_000
theta_mass_current = np.zeros(N_mcmc+1)
theta_distance_current = np.zeros(N_mcmc+1)
# use our prior as a guess
theta_mass_current[0] = 77.
theta_distance_current[0] = 1700.

# choose a value for width of normal distribution to get the step in height
# this is between the prior and the likelihood values
sigma_mass_mcmc = 0.1
sigma_distance_mcmc = 100


for i in tqdm(range(N_mcmc)):
    p_current = posterior(theta_mass_current[i], theta_distance_current[i]) # put current value in posterior equation
    dtheta_mass = np.random.normal(0,sigma_mass_mcmc) # randomally draw a value of theta to trial 
    dtheta_distance = np.random.normal(0,sigma_distance_mcmc) # randomally draw a value of theta to trial 
    theta_mass_proposed = theta_mass_current[i] + dtheta_mass #get new proposed theta (random theta + stepsize)
    theta_distance_proposed = theta_distance_current[i] + dtheta_distance #get new proposed theta (random theta + stepsize)
    p_proposed = posterior(theta_mass_proposed, theta_distance_proposed) # calculate posterior p for proposed theta
    
    #  keep this value if probability proposed theta greater than the current prob
    if p_proposed > p_current: 
        theta_mass_current[i+1] = theta_mass_proposed
        theta_distance_current[i+1] = theta_distance_proposed
    else:
    # if probability lower
    # use the ratios of probability to define probability of whether we move to that value or not
        p_new_move = np.exp(p_proposed-p_current)
        # generate random number for probability
        u_random = np.random.uniform(0,1)
        # if u_random < p_new_move, then accept, if not, reject
        if u_random <= p_new_move:
            theta_mass_current[i+1] = theta_mass_proposed
            theta_distance_current[i+1] = theta_distance_proposed
        else:
            theta_mass_current[i+1] = theta_mass_current[i]
            theta_distance_current[i+1] = theta_distance_current[i]

# get mean + std from mcmc generated samples
mean_mass_mcmc=np.mean(theta_mass_current)
median_mass_mcmc=np.median(theta_mass_current)
std_mass_mcmc=np.std(theta_mass_current, ddof=1)
mean_distance_mcmc=np.mean(theta_distance_current)
median_distance_mcmc=np.median(theta_distance_current)
std_distance_mcmc=np.std(theta_distance_current, ddof=1)

print('\nThe mean height from the MCMC is {:.2f} +/- {:.2f} cm'.format(mean_mass_mcmc,std_mass_mcmc))
print(median_mass_mcmc)
print(median_distance_mcmc)
```
