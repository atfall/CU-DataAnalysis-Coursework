# Probabilities

## <font color=#30a64d>Question</font>

You flip a coin 5 times. Use an appropriate probability function to estimate the probability that heads will **not** occur 3 times.  

### Answer

$B_{N, \theta} (\nu) = {N \choose \nu} \theta^\nu (1 - \theta)^{(N - \nu)}.$

```python
from scipy.stats import binom
theta = 0.5 #prob
N=5  # No of trials
nu=2 # No of successes

prob = binom.pmf(nu,N, theta)
print(1-prob)
```

### Answer

```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
theta = 0.7 #prob
N=17  # No of trials
x = np.arange(0,25,0.1)
mean, var, skew, kurt = binom.stats(N, theta, moments='mvsk')

y = st.norm.pdf(x, mean, var)
plt.plot(x,y)
```

## <font color=#30a64d> Question </font>

Plot a normal distribution in python with mean = 8.0 and standard deviation = 2.0 over an x range `[0,16]`. Do this both by writing your own function and by using in-built Python routines.  

Plot the CDF of the normal distribution using in-built python routines. Estimate the median.

### Answer

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

mean = 8.0
std = 2.0
x = np.arange(0, 16, 0.1)

y = st.norm.cdf(x, mean, std)
plt.plot(x,y)
```

## Question

Let $X$ be a discrete random variable with range $X={1,2,3,...}$. Suppose the probability mass function (PMF) of $X$ is given by

$P_X(k) = \dfrac{1}{2^k}$ for $k={1,2,3,...}$
 
1. Find $P(2<X\le 5)$.
2. Plot the CDF of $X$, $F_X(x)$ where
    $F_X(x) = \dfrac{2^k-1}{2^k}$

### Answer

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(1, 40, 1)

def px_f(k):
    return 1/(2**k)


def fx_f(k):
    return ((2**k)-1) / 2**k


def prob(l, u): # [l,u)
    val = 0
    for i in range(l,u,1):
        val += px_f(i)
    return val


#px = [px_f(i) for i in x]
#plt.plot(x, px)
#plt.show()

print(prob(3,6))

fx = [fx_f(i) for i in x]
plt.plot(x, fx)
plt.show()
```
